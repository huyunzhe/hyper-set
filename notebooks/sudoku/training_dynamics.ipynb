```python
import numpy as np
import pandas as pd

# Read the CSV file into a pandas DataFrame
file_path = "wandb_export_2025-04-28T00_31_00.963+08_00.csv"
df = pd.read_csv(file_path)

# Convert the DataFrame to a NumPy array
data_array = df.to_numpy()
```

```python
import matplotlib.pyplot as plt

x = np.arange(0, 200)
zero = np.zeros(200)
tf = data_array[:, -3]
ours = data_array[:, -6]
```

```python
plt.rcParams.update(
    {
        "text.usetex": True,
        "font.size": 10,
        "axes.labelsize": 10,
        "xtick.labelsize": 10,
        "ytick.labelsize": 10,
        "legend.fontsize": 8,
        "font.family": "Times New Roman",
        "grid.alpha": 0.3,
        "axes.linewidth": 1.0,
        "axes.grid": True,
        "grid.linestyle": "--",
    }
)

fig, ax = plt.subplots(figsize=(3.3, 2.2))

ax.set_xlabel("Training Epochs")
ax.set_ylabel(r"Board Accuracy (\%)")

ax.plot(ours, "-", label="Ours", markersize=4)
ax.plot(tf, "-", label="Transformer", markersize=4)
ax.plot(zero, "-", label="Energy Transformer", markersize=4)
ax.plot(zero, "-", label="CRATE", markersize=4)
ax.grid(True)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.legend(loc="lower right", frameon=False)
plt.tight_layout()
# plt.savefig("sudoku_training_dynamics.pdf", bbox_inches="tight", dpi=600)
plt.show()
```
